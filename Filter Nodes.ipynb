```python
N = 5
```

```python
coord = pd.read_csv("data/dataviz.csv",index_col=0)
display(coord.head())
```

```python
node_usage = pd.read_csv("data/nodes_usage_delivery_ids.csv",index_col=0)
display(node_usage.shape)
display(node_usage.head(20))
```

```python
node_usage.columns
```

```python
def filterN(N):
    nodes_below_thres = node_usage[node_usage.tot_usage <= N].node_id
    #nodes_below_thres= nodes_below_thres[['node_id','lat','lon']]
    print('Num nodes below threshold N=',N,' is: ',nodes_below_thres.shape[0])
    node_usage_filtered = node_usage[node_usage.tot_usage > N]
    print('Node usage filtered shape: ',node_usage_filtered.shape)
    namefile= 'nodes_usage_delivery_filtered_N'+str(N)+'.csv'
    node_usage_filtered.to_csv('data/'+namefile)
    return node_usage_filtered, nodes_below_thres
```

```python
node_usage_filter,nodes_below_threshold= filterN(N)
```

```python
nodes_below_threshold.head(2)
```

```python
# change road type from string to list
from ast import literal_eval
coord['road'] = coord.road.apply(lambda x: literal_eval(x))
coord.dtypes
```

```python
new_coord =coord.copy()
new_coord.head(2)
```

```python
def inRoad(road_list,node_id):
    if(node_id in road_list):
        #print('Removed node_id: ',node_id)
        road_list.remove(node_id)
    return road_list        
```

```python
count=0
for node in nodes_below_threshold:
    #print(n_coord)
    print(count,'/',nodes_below_threshold.shape[0])
    new_coord['road'] = new_coord.road.apply(lambda x: inRoad(x,node))
    count=count+1
```
