```python
import pandas as pd
import osmapi
import time
```

```python
api = osmapi.OsmApi()
```

```python
df = pd.read_csv("data/dataviz.csv")
print(df)
```

```python
#pickup_points = df[['plat', 'plng']].drop_duplicates().reset_index(drop = True)
```

```python
unique_nodes = set()
for ids in df['road']:
    id_list = ids.split(', ')
    id_list[0] = id_list[0][1:]
    id_list[-1] = id_list[-1][:-1]
    for i in id_list:
        if i not in unique_nodes:
            unique_nodes.add(i)
#print(unique_nodes)
print(len(unique_nodes))
```

```python
#unique_nodes = {2223091605, 311463563}
```

```python
f = open("node_id_to_coordinates.csv","w+")
f.write('node_id,lat,lon\n')

total = len(unique_nodes)
count = 1
error_count = 0
for node_id in unique_nodes:
    print(str(count) + '/' + str(total))
    count += 1
    time.sleep(0.1)
    try:
        node = api.NodeGet(node_id)
        f.write(str(node_id) + ', ' + str(node['lat']) + ', ' + str(node['lon']) + '\n')
    except:
        print('ERROR ON ' + str(node_id))
        error_count += 1
 
f.close()
```

## Count Nodes in a Road

```python
# Analyse routes (list of nodes) taken be drivers
df['numNodes'] = df.road.apply(lambda x: len(x))
display(df.head())
```

```python
# According to data for a delivery a driver passes avg of 2380 nodes.
df.numNodes.describe()
```

Max num of nodes in the road is 7429 whereas min is 159.

```python
# check if in a node there is duplicated node
df['duplicated_nodes_in_road'] = df.road.apply(lambda x: len(x) == len(set(x)))
display(df[df.duplicated_nodes_in_road==True])  # There is none -- santiy check is correct
display(df.head())
df.drop(columns=['duplicated_nodes_in_road'], inplace=True)
```

```python
df.dtypes
```

```python
# change road type from string to list
from ast import literal_eval
df['road_list'] = df.road.apply(lambda x: literal_eval(x))
df.dtypes
```

```python
df.head()
df.drop(columns=['road_list'])
```

```python
df.to_csv('dataviz_numNodes.csv')
```

```python
# given a node find the all routes using this node
# which deliveries use this node? return the ones id
def findRoutes(node):
    #print('Node id: ',node)
    id_deliveries = []
    for idx, d in df.iterrows() :
        if(node in d.road_list):
            id_deliveries.append(idx)
    return  id_deliveries
```

```python
node_id = 311463563
res = findRoutes(node_id)
print('Node ',node_id, ' used ', len(res),' different deliveries.')
print('The deliveries id are: ',res)
```
