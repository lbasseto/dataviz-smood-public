```python
import pandas as pd
import re
```

```python
node_id_to_coord = pd.read_csv("data/nodes_usage_delivery_ids.csv")
node_id_to_coord.head()
```

```python
node_id_to_coord_dict = {r['node_id']: str(r['lat'])[:9] + '/' + str(r['lon'])[:8] for _, r in node_id_to_coord.iterrows()}
node_id_to_coord_dict
```

```python
df = pd.read_csv("data/dataviz_filtered_N5.csv")
df.head()
```

```python
def str_of_ids_to_coords(s):
    id_list = s.split(', ')
    id_list[0] = id_list[0].replace('[', '')
    id_list[-1] = id_list[-1].replace(']', '')
    string = str([node_id_to_coord_dict.get(int(node_id)) for node_id in id_list])
    return string.replace(' ', '').replace('\'', '').replace('[', '').replace('', '')

str_of_ids_to_coords('[311463563, 2940544168, 311463564]')
```

```python
df['road'] = df['road'].apply(str_of_ids_to_coords)
df['road']
```

```python
df.to_csv('data/dataviz_filtered_coords_N5.csv')
```
